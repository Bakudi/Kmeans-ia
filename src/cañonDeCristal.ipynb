```python
import pandas as pd

# Cargar la base de datos
df = pd.read_csv("Pokemon.csv")

# Filtrar columnas relevantes
df_stats = df[['Name', 'Sp. Atk', 'Defense', 'Sp. Def']]
df_stats = df_stats.dropna()

```

```python
# Definir una métrica combinada (mayor es mejor para cañones de cristal)
df_stats['GlassCannonScore'] = df_stats['Sp. Atk'] - (df_stats['Defense'] + df_stats['Sp. Def']) / 2
```

```python
# Ordenar por la métrica
top_glass_cannons = df_stats.sort_values(by='GlassCannonScore', ascending=False)

# Mostrar los top 20
top_glass_cannons.head(30)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.scatterplot(data=df_stats, x='Defense', y='Sp. Atk', hue='GlassCannonScore', palette='coolwarm')
plt.title('Distribución de Cañones de Cristal (Sp. Atk vs Defense)')
plt.xlabel('Defensa')
plt.ylabel('Ataque Especial')
plt.grid(True)
plt.show()
```
